### Silme Yöntemleri

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame(
            {"V1" : V1,
             "V2" : V2,
             "V3" : V3
            }
)
df
```

```python
df.dropna() #en az 1 eksik olan yok oldu tamamen değil
```

```python
df.dropna(how = "all") #hepsi aynı anda eksik  ise sil
```

```python

df.dropna(axis = 1) #en az 1 eksik değere sahip değeri sil;
```

```python
df.dropna(axis =1 ,how= "all") 
#tüm değerleri boş olanı sil demek
```

```python
df.dropna(axis =1 ,how= "all",inplace =True)  #ile inplace ile kalıcı olarak silmeyi yaparız
```

### Değer Atama Yöntemi

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame(
            {"V1" : V1,
             "V2" : V2,
             "V3" : V3
            }
)
df
```

```python
#Sayısal değişkenlerde atama
```

```python
df["V1"].fillna(df["V1"].mean()) #ortalaması ile dolduruzu
```

```python
df.apply(lambda x: x.fillna(x.mean()),axis=0)#her gözlem birimini kendi ortalaması ile dordurttuk
```

```python
df.fillna(df.mean()[:]) #bu da aynıı işlemi yapar
```

```python
df.fillna(df.mean()["V1":"V2"])
```

```python
df["V3"].fillna(df["V3"].median())#mediyanı ile doldurur
```

```python
df.where(pd.notna(df),df.mean(),axis = "columns")#eksik değer yerine o değerlerin ort doldurur
```

### Kategorik değişken kırılımında değer atama

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
V4 = np.array(["IT","IT","IK","IK","IK","IK","IK","IT","IT"])
df = pd.DataFrame(
            {"maas" : V1,
             "V2" : V2,
             "V3" : V3,
             "departman" : V4
            }
)
df
```

```python
df.groupby("departman")["maas"].mean()#departmana göre ortalama
```

```python
df["maas"].fillna(df.groupby("departman")["maas"].transform("mean"))
#maas değişkeni eksikliğini bulunduğu departmana uygun bir ortalama ile doldurulur 
```

### Kategorik değişkenler için eksik değer atama

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])

V4 = np.array(["IT",np.NaN,"IK","IK","IK","IK","IK","IT","IT"],dtype=object)
df = pd.DataFrame(
            {"maas" : V1,
             
             "departman" : V4
            }
)
df
```

```python
df["departman"].fillna(df["departman"].mode()[0]) #modunu aldık ve sondaki 0 da stingolarak IK almamızı sağladı 
```

```python
df["departman"].fillna(method = "bfill") #sonraki değer ile doldurur
```

```python
df["departman"].fillna(method = "ffill") #kendinde önceki değer ile doldurdu
```
